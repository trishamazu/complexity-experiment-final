```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Load the CSVs
csv1_path = "/home/wallacelab/complexity-final/Experiments/Data/bistable_2afc_choices.csv"  # Replace with the path to the first CSV
csv2_path = "/home/wallacelab/complexity-final/Embeddings/CLIP-CBA/Bistable/crossval_cba_predicted_bistable_complexity_scores.csv"  # Replace with the path to the second CSV

# Read the CSVs into DataFrames
csv1 = pd.read_csv(csv1_path)
csv2 = pd.read_csv(csv2_path)

# Extract the relevant columns
column1 = csv1['Proportion of Participant Choice']
column2 = csv2['Predicted Complexity Score']

# Calculate the differences between every two rows
column1_diff = column1.iloc[::2].values - column1.iloc[1::2].values
column2_diff = column2.iloc[::2].values - column2.iloc[1::2].values

# Ensure the lengths match
if len(column1_diff) != len(column2_diff):
    raise ValueError("The number of differences in the two datasets do not match!")

# Compute Spearman correlation
spearman_corr, p_value = stats.spearmanr(column1_diff, column2_diff)

# Generate a scatter plot with a line of best fit
plt.figure(figsize=(8, 6))
plt.scatter(column1_diff, column2_diff, color='blue', label='Data Points')

# Line of best fit
z = np.polyfit(column1_diff, column2_diff, 1)  # Linear fit
p = np.poly1d(z)
plt.plot(column1_diff, p(column1_diff), color='red', linestyle='--', label='Line of Best Fit')

# Add annotations
plt.title("Correlation Between Differences", fontsize=16)
plt.xlabel("Difference in Number of Participant Choice", fontsize=14)
plt.ylabel("Difference in Predicted Complexity Score", fontsize=14)
plt.legend(fontsize=12, loc='upper left')
plt.grid(True)

# Add correlation information
plt.annotate(f"Spearman r = {spearman_corr:.2f}\nP-value = {p_value:.2e}",
             xy=(0.05, 0.84), xycoords='axes fraction',
             fontsize=12, ha='left', va='top',
             bbox=dict(boxstyle='round,pad=0.3', edgecolor='gray', facecolor='white'))

print(f"Spearman Correlation: {spearman_corr:.2f}")
print(f"P-value: {p_value:.2e}")

# Show the plot
plt.show()
```
