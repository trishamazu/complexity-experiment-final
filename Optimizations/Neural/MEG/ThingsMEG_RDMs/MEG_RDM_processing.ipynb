```python
path = "./validation-pairwise_decoding_RDM1854.mat"

import h5py
import numpy as np
import matplotlib.pyplot as plt

with h5py.File(path, 'r') as f:
    data = f['mat']
    data = np.array(data)
    print(data.shape)
```

```python
rdm = np.mean(data, axis=0)
rdm = np.nan_to_num(rdm)
```

```python
rdm[100].min()
```

```python
overall_dist = np.mean(rdm, axis=(1,2))
print(overall_dist.shape)
```

```python
overall_dist
plt.plot(overall_dist)
plt.xlabel('Time')
plt.ylabel('Mean RDM distance')
plt.title('THINGsMEG RDM Mean Distance over Time')
plt.show()
```
